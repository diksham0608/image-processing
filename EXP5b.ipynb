```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = "/content/diksha.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (600, 400))  # Resize for better visualization

# Define Prewitt Kernels
prewitt_x = np.array([[1, 0, -1],
                       [1, 0, -1],
                       [1, 0, -1]], dtype=np.float32)  # Horizontal edges

prewitt_y = np.array([[1,  1,  1],
                       [0,  0,  0],
                       [-1, -1, -1]], dtype=np.float32)  # Vertical edges

# Apply filter2D for edge detection
prewitt_x_edge = cv2.filter2D(image, -1, prewitt_x)
prewitt_y_edge = cv2.filter2D(image, -1, prewitt_y)

# Compute combined gradient magnitude
prewitt_combined = np.sqrt(np.square(prewitt_x_edge) + np.square(prewitt_y_edge))
prewitt_combined = np.uint8(prewitt_combined)

# Display results
titles = ["Original", "Prewitt X", "Prewitt Y", "Prewitt Combined"]
images = [image, prewitt_x_edge, prewitt_y_edge, prewitt_combined]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
