```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "/content/diksha.jpg"  # Replace with your image
image = cv2.imread(image_path)
image = cv2.resize(image, (600, 400))  # Resize for better visualization

# Convert to grayscale (optional)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define sharpening kernels
sharpening_kernels = {
    "Basic Sharpen": np.array([[0, -1,  0],
                               [-1,  5, -1],
                               [0, -1,  0]], dtype=np.float32),

    "Strong Sharpen": np.array([[-1, -1, -1],
                                [-1,  9, -1],
                                [-1, -1, -1]], dtype=np.float32),

    "Excessive Sharpen": np.array([[-1, -2, -1],
                                   [-2,  13, -2],
                                   [-1, -2, -1]], dtype=np.float32)
}

# Apply different sharpening filters
sharpened_images = {name: cv2.filter2D(image, -1, kernel) for name, kernel in sharpening_kernels.items()}

# Display results using Matplotlib
titles = ["Original"] + list(sharpened_images.keys())
images = [image] + list(sharpened_images.values())

plt.figure(figsize=(14, 8))
for i in range(len(images)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
