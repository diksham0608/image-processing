```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "/content/diksha.jpg"  # Replace with your image file
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (600, 400))  # Resize for better visualization

# Apply Gradient Operators

# 1. Sobel Operator (X and Y)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges
sobel_combined = cv2.magnitude(sobel_x, sobel_y)  # Magnitude of gradients

# 2. Prewitt Operator (X and Y)
prewitt_x = cv2.filter2D(image, -1, np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]]))  # Horizontal
prewitt_y = cv2.filter2D(image, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))  # Vertical

# 3. Laplacian Operator (Second-order derivative)
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# 4. Scharr Operator (More precise than Sobel)
scharr_x = cv2.Scharr(image, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(image, cv2.CV_64F, 0, 1)
scharr_combined = cv2.magnitude(scharr_x, scharr_y)

# Display Results
titles = ["Original", "Sobel X", "Sobel Y", "Sobel Combined",
          "Prewitt X", "Prewitt Y", "Laplacian", "Scharr Combined"]
images = [image, sobel_x, sobel_y, sobel_combined,
          prewitt_x, prewitt_y, laplacian, scharr_combined]

plt.figure(figsize=(12, 10))
for i in range(len(images)):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
